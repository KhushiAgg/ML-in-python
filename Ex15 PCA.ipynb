```python
df2 = df.copy()
df2.ExerciseAngina.replace(
    {
        'N' : 0,
        'Y' : 1
    },
    inplace=True)
df2.RestingECG.replace(
    {
        'Normal' : 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True)
df2.ST_Slope.replace(
    {
        'Up' : 1,
        'Flat' : 2,
        'Down' : 3
    },
    inplace=True)
df2.head()
```

```python
df3 = pd.get_dummies(df2, drop_first = True)
df3.head()
```

```python
X = df3.drop(['HeartDisease'], axis = 1)
y = df3.HeartDisease
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
svm = SVC()
lr = LogisticRegression()
rfc = RandomForestClassifier()
```

```python
model_params = {
    'svm' : {
        'model' : svm,
        'params' : {
            'C' : [1, 10, 20],
            'kernel' : ['rbf', 'linear']
        }
    },
    'random_forest' : {
        'model' : rfc,
        'params' : {
            'n_estimators' : [1, 5, 10]
        }
    },
    'logistic_regression' : {
        'model' : lr,
        'params' : {
            'C' : [1, 5, 10]
        }
    }
}
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.3)

scores = []

for model_name, mp in model_params.items():
    clf = GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    clf.fit(X, y)
    scores.append({
        'model' : model_name,
        'best_score' : clf.best_score_,
        'best_params' : clf.best_params_
    })
```

```python
import pandas as pd
df4 = pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
df4
```

```python
from sklearn. decomposition import PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(X_scaled)
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.3, random_state=30)
```

```python
svm.fit(X_train_pca, y_train)
svm.score(X_test_pca, y_test)
```

```python
rfc.fit(X_train_pca, y_train)
rfc.score(X_test_pca, y_test)
```
