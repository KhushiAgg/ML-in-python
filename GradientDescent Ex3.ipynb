```python
def gradient_descent(x, y):
    m_cur = b_cur = 0
    iterations = 1000
    n = len(x)
    alpha = 0.0002
    
    cost_previous = 0
    
    for i in range(iterations):
        y_predicted = m_cur * x + b_cur
        cost_func = (1/n) * sum(x*(y - y_predicted))
        mderivative = (2/n) * sum(x*(y - y_predicted))
        bderivative = (2/n) * sum(y - y_predicted)
        m_cur = m_cur - alpha * mderivative
        b_cur = b_cur - alpha * bderivative
#         if math.isclose(cost_func, cost_previous, rel_tol = 1e-20, abs_tol = 0.0):
#             break
#         cost_previous = cost_func
        print("m {}, b {}, cost {}, iterations {}".format(m_cur, b_cur, cost_func, i))
```
