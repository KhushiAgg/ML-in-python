```python
# Removing outliers using z score
```

```python
df['z_score_RestingBP'] = (df.RestingBP - df.RestingBP.mean())/df.RestingBP.std()
df['z_score_Cholesterol'] = (df.Cholesterol - df.Cholesterol.mean())/df.Cholesterol.std()
df['z_score_MaxHR'] = (df.MaxHR - df.MaxHR.mean())/df.MaxHR.std()
df['z_score_Oldpeak'] = (df.Oldpeak - df.Oldpeak.mean())/df.Oldpeak.std()
df1 = df[(df.z_score_RestingBP > -3) & (df.z_score_RestingBP <3)]
df1 = df[(df.z_score_Cholesterol > -3) & (df.z_score_Cholesterol <3)]
df1 = df[(df.z_score_MaxHR > -3) & (df.z_score_MaxHR <3)]
df1 = df[(df.z_score_Oldpeak > -3) & (df.z_score_Oldpeak <3)]
df1.head()
```

```python
df1.drop(['z_score_RestingBP', 'z_score_Cholesterol', 'z_score_MaxHR', 'z_score_Oldpeak'], axis = 1, inplace = True)
df1.shape
```

```python
df1.ChestPainType.unique()
```

```python
df1.RestingECG.unique()
```

```python
df1.ExerciseAngina.unique()
```

```python
df1.ST_Slope.unique()
```

```python
df2 = df1.copy()
df2.RestingECG.replace(
    {
        'Normal' : 1,
        'ST' : 2,
        'LVH' : 3
    },
    inplace = True
)
df2.ExerciseAngina.replace(
    {
        'N':0,
        'Y':1
    },
    inplace = True
)
df2.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace = True
)
df2.head()
```

```python
df2 = pd.get_dummies(df2, drop_first = True)
df2.head()
```

```python
df2.shape
```

```python
X = df2.drop(['HeartDisease'], axis = 1)
y = df2.HeartDisease
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.3, random_state = 30)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

svm = cross_val_score(SVC(), X_scaled, y, cv=5)
svm.mean()
```

```python
from sklearn.ensemble import BaggingClassifier

bag_model = BaggingClassifier(
    base_estimator=SVC(),
    n_estimators = 100,
    max_samples = 0.8,
    oob_score = True,
    random_state = 0
)
scores = cross_val_score(bag_model, X_scaled, y, cv=5)
scores.mean()
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = cross_val_score(DecisionTreeClassifier(random_state=0), X_scaled, y, cv=5)
tree.mean()
```

```python
bag_model_2 = BaggingClassifier(
    base_estimator=DecisionTreeClassifier(),
    n_estimators = 100,
    max_samples = 0.8,
    oob_score = True,
    random_state = 0
)
scores = cross_val_score(bag_model_2, X_scaled, y, cv=5)
scores.mean()
```

```python
from sklearn.ensemble import RandomForestClassifier
rfr = RandomForestClassifier()
rfr.fit(X_train, y_train)
rfr.score(X_test, y_test)
```
